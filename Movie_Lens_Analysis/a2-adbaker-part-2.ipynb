## Part 2: Data Cleansing and Plotting
A. Use the Benford's Law formula to plot a line graph depicting the probability distribution of P(d), where d is the leading digit in a data set.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot') 

def prob_d(d):
    '''Accepts digit d as input, and returns the probability of d occuring
    in a dataset as defined by Benford\'s Law'''
    return np.log10(1 + (1 / d))

x = range(1, 10)
y = [prob_d(d) for d in x]

fig, ax = plt.subplots(figsize=(15,7))
plt.title('Benford\'s Law')
plt.xlabel('d')
plt.ylabel('P(d)')
plt.plot(x, y, color = '#ff0066')
plt.show()
```

B. The Wikipedia article https://en.wikipedia.org/wiki/List_of_tallest_buildings_and_structures has a table of heights of tallest buildings / structures in the world. Using Pandas read_html, read the table under the section “Tallest Structure by Category”. Note that read_html will read all the tables on a page and return a list. Plot the frequency of occurrence of the leading digit in height (feet) on the same figure of as the bar graph.

```python
tallest_structures = pd.read_html('https://en.wikipedia.org/wiki/List_of_tallest_buildings_and_structures', header = 0)
tall_struct_df = tallest_structures[3]

def get_first_dig(n):
    'Accepts a number n and returns the first digit of n'
    while n > 9:
        n = n // 10
    return int(n)

# creating a second dataframe to group the observed first digits
tall_struct_df['First Digit (feet)'] = [get_first_dig(n) for n in tall_struct_df['Height (feet)']]
digits_df = pd.DataFrame(tall_struct_df['First Digit (feet)'].value_counts()).sort_index()
digits_df['P(d)'] = [i / np.sum(digits_df['First Digit (feet)']) for i in digits_df['First Digit (feet)']]

# plotting data from Wikipedia and Benford's law together
fig, ax = plt.subplots(figsize=(15,7))
plt.title('Comparing Benford\'s Law Against Building Height Data')
plt.xlabel('d')
plt.ylabel('P(d)')
plt.plot(x, y, color = '#ff0066', label = 'Benford\'s Law') # x and y defined above in part A
plt.bar(digits_df.index, digits_df['P(d)'], color = '#7E36F2', label = 'Tallest Buildings Data (Feet)')
plt.xticks(np.arange(1,10))
plt.legend(loc='upper right')
plt.show()
```

C. Repeat part B. using height in meters

```python
import re

# creating a second dataframe to clean the 'meters' column data and then group the observed first digits
tall_struct_df['Height (meters) (clean)'] = tall_struct_df['Height (meters)'].apply(lambda v: re.sub(r'\[\d\]', '', v))
tall_struct_df['First Digit (meters)'] = [get_first_dig(float(n)) for n in tall_struct_df['Height (meters) (clean)']]
digits_df_met = pd.DataFrame(tall_struct_df['First Digit (meters)'].value_counts()).sort_index()
digits_df_met['P(d)'] = [i / np.sum(digits_df_met['First Digit (meters)']) for i in digits_df_met['First Digit (meters)']]

# plotting data on building height (in meters and feet) against Benford's Law
fig, ax = plt.subplots(figsize=(15,7))
plt.title('Comparing Benford\'s Law Against Building Height Data')
plt.xlabel('d')
plt.ylabel('P(d)')
plt.plot(x, y, color = '#ff0066', label = 'Benford\'s Law')         # from part A
plt.bar(digits_df.index, digits_df['P(d)'], color = '#7E36F2', width = .5,
        label = 'Tallest Buildings Data (Feet)')                    # from part B
plt.bar(digits_df_met.index, digits_df_met['P(d)'], color = '#FCDC19', align = 'edge', width = .5,
        label = 'Tallest Buildings Data (Meters)')                  # second bar chart for meters
plt.xticks(np.arange(1,10))
plt.legend(loc='upper right')
plt.show()
```
